```python
sns.set_style("white")
sns.set_context("poster",font_scale = .7)

palette = ["#007f7b", "#6ab097", "#f2aa86", "#e7353b", "#ffa26e", "#b09e6d", "#ff7000", "#c11528",
           "#b20323", "#cfcfcf", "#ffeb3e", "#ffd700", "#424542", "#252525"]
```

```python
plt.figure(figsize=(10, 6)) 
ax = sns.countplot(x='gender', hue='Churn', data=data, palette=palette[2:])


plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Gender vs. Churn")


total = len(data)
for p in ax.patches:
    count = p.get_height()
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.text(x, y, f'{count}', ha='center', va='bottom')


plt.legend(title="Churn", loc='upper right', labels=['No', 'Yes'])
plt.show()
```
